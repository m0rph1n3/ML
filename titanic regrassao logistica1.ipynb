```python
def input_Age_For_Nulls(pCols):
    myAge = pCols[0]
    myClass = pCols[1]
    
    if pd.isnull (myAge):
        if myClass == 1:
            return 37
        elif myClass == 2:
            return 29
        else:
            return 24
    else:
        return myAge
```
